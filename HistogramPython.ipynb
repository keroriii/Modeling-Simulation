```python
import matplotlib.pyplot as plt
import numpy as np

# Data array
data = np.array([20.00, 49.67, 92.00, 56.53, 73.07, 78.40, 68.13, 60.27, 88.13, 67.60,
                 78.00, 65.87, 53.33, 65.47, 47.20, 65.20, 53.60, 80.53, 60.53, 75.33,
                 66.20, 50.40, 69.13, 56.67, 60.53, 63.73, 56.27, 51.20, 66.67, 47.67,
                 16.67, 50.27, 37.33, 81.20, 80.40, 59.73, 63.07, 58.80, 66.67, 65.07,
                 65.60, 62.40, 81.73, 64.80])

# Create histogram
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data, bins=10, color='skyblue', edgecolor='white')

# Add title and labels
plt.title('Distribution of Values (MDTERM)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add mean line
mean_val = np.mean(data)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, 
            label=f'Mean = {mean_val:.2f}')

# Set x-axis limits
plt.xlim(0, 100)

# Add legend and grid
plt.legend()
plt.grid(axis='y', alpha=0.5)

plt.show()
```
